# Run the Python file containing the SFGPL examples

```python
import subprocess
import os
from SFGPL import *
```

## Set default directories

```python
PY_SAMPLES_DIR="py/samples/"
OUT_FILES_DIR="out/"
```

## Set file settings

### Enumerate the Python and output JSON files included in the project.

```python
file_datas={
    "sample01":{
        "py":PY_SAMPLES_DIR+"sample01.py",
        "json":OUT_FILES_DIR+"sample01.json"
    },
    "sample02":{
        "py":PY_SAMPLES_DIR+"sample02.py",
        "json":OUT_FILES_DIR+"sample02.json"
    },
}
```

### List files to be excluded.

```python
exclusion_key_list=[
    "sample02"
]
```

## Run Python files

```python
run_pys=[file_datas[key]["py"] for key in file_datas if key not in exclusion_key_list]
out_jsons=[file_datas[key]["json"] for key in file_datas if key not in exclusion_key_list]
```

```python
# List of Python files to be executed
for p in run_pys:
    print(p)
```

# Create a JSON file of the integrated SFGPL corpus

```python
sc=SFGPLCorpus()
for json_path in out_jsons:
    sc+=SFGPLCorpus.readJson(json_path)
```

```python
print(len(sc))
```

```python
sc.saveJson(OUT_FILES_DIR+"sample.json")
```

```python
sc.saveStructureJSON(OUT_FILES_DIR+"structured_sample.json")
```

```python
sc_wc=sc.corpusWordCount()
print(len(sc_wc),sum([x[1] for x in sc_wc.items()]))
```
